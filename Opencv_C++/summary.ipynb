### Functions

Read image: `Mat img = imread(path); `

Show image always: `imshow("Image", img); waitKey(0); // wait infinity`

Load video: `VideoCapture cap(path); While Loop; cap.read(img); imshow("Image", img); waitKey(1);`

Read camera: `VideoCapture cap(0); While Loop; cap.read(img);imshow("Image", img); waitKey(1);`

Converts an image from one color space to another: `cvtColor()`

Blur: `GaussianBlur()`

Finds edges: `Canny()`

`dilate()`, `erode()`, `resize()`

Draw: `circle()`, `rectangle()`, `line`, `putText`

Warp:

1. define source and destination points `point2f (float _x, float _y)`

2. get matrix `matrix = getPerspectiveTransform(src, dst);`

3. `warpPerspective(img, imgWarp, matrix, Point(w,h));`

Named window & trackbar:

`namedWindow("Trackbars", 0);
 resizeWindow("Trackbars", 500, 200);
 createTrackbar("Hue Min", "Trackbars", &hmin, 179);`

Filter color:

Convert RGB to HSV (Hue, Saturation, Value)

`cvtColor(img, imgHSV, COLOR_BGR2HSV);
 inRange(imgHSV, lower, upper, mask);`
 
Find contours:
`findContours(imgDil, contours, hierarchy, RETR_EXTERNAL, CHAIN_APPROX_SIMPLE);`

Create bounding rectangle:

`vector<Rect> boundRect(contours.size());
 float peri = arcLength(contours[i], true);
 approxPolyDP(contours[i], conPoly[i], 0.02 * peri, true);
 boundRect[i] = boundingRect(conPoly[i]);`

Draw contours:
`rectangle(img, boundRect[i].tl(), boundRect[i].br(), Scalar(0, 255, 0), 5);`

Face detector:

`CascadeClassifier faceCascade;
 faceCascade.load("Resources/haarcascade_frontalface_default.xml");
 vector<Rect> faces;
 faceCascade.detectMultiScale(img, faces, 1.1, 10);`
